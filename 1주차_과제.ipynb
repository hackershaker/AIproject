## 언어

  자동번역(또는 기계번역)이라고 불리는 기술은 크게 규칙기반 기계번역(RBMT, Rule-Based Machine Translation)과 통계기반 기계번역(SMT, Statistical Machine Translation), 이를 합한 하이브리드(RBMT+SMT) 기반 기계번역으로 구분된다.

  규칙기반 기계번역은 문법에 중점을 두고 번역하는 방식이다. 이 방식은 문법에 맞는 글엥 적용될 경우 정확성이 높지만, 문법이 약간 어긋나면 번역의 신뢰도가 떨어진다. 통계기반 기계번역은 컴퓨터가 읽을 수 있는 형태로 저장된 일정 규모 이상의 언어 자료인 '말뭉치(corpus)'를 미리 번역 엔진에 입력해 놓은 뒤 통계적으로 봤을 때 번역을 요청한 문장 내에 단어나 구문과 가장 비슷하다고 판단한 결과를 내놓는 방식이다. 동의어의 구별, 언어별로 서순이 다른 점이 잘 적용되지 않는 문제가 있다. 
  
  현재 구글에서 서비스하고 있는 구글 번역기는 딥러닝과 신경망 기술이 적용된 신경망 기계번역(GNMT, Google’s Neural Machine Translation)을 사용한다. 머신러닝을 이용하여 전체 문맥을 파악하고 단어의 의미 차이 등을 반영하여 훨씬 자연스러운 문장 번역이 가능하다. 하지만 구글 내부에서는 고유명사와 희귀용어의 오역, 문맥 또는 문단을 고려하지 않은 문장 번역 등의 개선 과정이 필요하다고 말했다. 과연 인공지능과 결합된 번역 시스템이 인간 번역가를 뛰어넘을 날이 올지 지켜봐야 할 일이다.

## 음성

  대부분의 음성인식 서비스는 딥러닝에 기반한다. 딥러닝 기반 음성인식 프로세스는 다음과 같다. 먼저 언어의 소리단위(우리나라 자음 모음 등)를 딥러닝으로 학습시켜 음향모델을 만든다. 이후 언어 모델, 어휘사전, 잡음 규칙 등을 적용시켜 입력된 음성정보를 디코딩하여 인식 결과를 출력한다. 인공신경망에는 단순히 뉴런을 연결시킨 구조만 있어 과거의 정보가 필요할 때 참고하지 못하는 문제가 있다. 
  
  Saltlux의 음성인식 엔진은 LSTM(Long Short-Term Memory)기술을 적용하여 음성인식 성능을 높였다.  LSTM은 입력,출력,망각 게이트가 존재하여 필요한 기간 동안 이를 보존하기 위해 학습하고, 필요할 때마다 이를 추출하여 결과를 출력할 때에 사용한다.이는 입력된 문장 내의 단어가 어떤 뜻으로 쓰였는지 지난 문장의 문맥을 참고하여 보다 정확한 음성인식을 이루어낼 수 있다.


## 이미지

  스포츠를 좋아하는 사람들은 대부분 보는 것이 경기 다음 날 올라온 하이라이트 영상이다. NC AIlab에서는 이미지 인식 기술을 활용하여 야구 경기 하이라이트 클립을 자동으로 생성한다. 먼저 야구 중계 중 일어났던 사건들을 차례대로 기록한다. 이벤트, 각 팀의 점수, 이닝, 볼, 스트라이크, 아웃, 주자, 투구수, 투수, 타자, 해당 순번에 대한 텍스트 등으로 경기를 구별시켜 놓는 것이다. 이러한 과정을 거치면 특정 키워드를 이용하여 원하는 구간을 찾을 수 있다. 예컨대 {이닝: 5T, 타자: 나성범, 이벤트: 안타}와 같은 식이다. 그러나 여기에 영상의 시간 정보는 대응되어 있지 않기 때문에 기록 정보들을 영상 시점과 연결시키는 태깅(Tagging) 작업을 진행한다. 이때 태그 정보는 화면에 나타나는 중계 정보 자막이 나타나 있는 경우 이를 인식하여 생성한다. 원하는 하이라이트 영상을 찾으려면 특정 정보를 입력하여 검색하고 이와 태그되어 있는 구간을 찾을 수 있는 것이다. 
  
  이렇게 추출해 낸 영상에는 불필요한 정보들이 앞뒤로 붙어있는 경우가 있다. 이때는 프레임 간의 특정 정보에 대한 변화량을 산출하여 장면이 전환되는 시점을 찾고, 야구 경기와 관련된 장면(투구 등)과 경기 외적인 부분(광고 등)을 인공지능을 이용한 카테고리 분류 기술을 통해 잘라낼 수 있다. 이러한 과정을 통해 더 정교한 하이라이트 생성이 가능하다.
 
  벤치클리어링 같은 특별한 하이라이트 장면의 경우 아직 인식하여 클립으로 생성할 수 없다는 단점이 있다. 하이라이트 생성 기술이 더욱 고도화되고 발전한다면 엄청난 편리함을 가져다 줄 것이다.

## 자율주행

  자율주행에서 가장 중요한 점 중 하나는 실시간으로 변하는 도로의 상황에 잘 대응할 수 있는 인공지능을 만드는 것이다. 알파벳(Alphabet) 산하 자율주행 스타트업인 웨이모에서는 이러한 문제를 해결하기 위하여 인구 기반 트레이닝(population-based training: PBT)을 자율주행차에 적용하였다.  차량이 마주하는 상황이 바뀔 때마다 인공지능을 재학습시키고 자율주행 알고리즘을 재조정해야 한다. 이 과정에서 적절한 속성과 매개변수를 찾아 심층신경망을 얼마나 빠르고 효율적으로 구성하느냐가 핵심이라고 할 수 있다. PBT는 생물학적 진화에서 영감을 받은 알고리즘으로 주어진 업무를 가장 잘 수행하는 파라미터를 채택하여 특정 작업에 대한 학습 알고리즘과 매개 변수 선택 속도를 높인다. 웨이모는 PBT 기술로 인하여 학습 시간과 컴퓨터 자원을 각각 50%씩 절약하였다고 발표하였다.

  미국자동차기술학회(Society of Automotive Engineers, SAE)은 자동차 자율주행 기술을 6단계로 분류했다. 그 중 마지막 단계인 5단계는 완전 자동화 단계로 모든 주행 상황에서 운전자의 개입이 불필요하며 운전자 없이 주행 가능한 완전한 단계이다. 이 단계를 위하여 인공지능 알고리즘은 어떠한 도로 상황을 마주치더라도 판단을 내릴 수 있도록 혁신적인 발전이 필요할 것이다. 미래에는 운전자 없이 도로를 달리는 자동차가 당연한 시대가 오리라 기대한다.
